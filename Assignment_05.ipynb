```python
Assignment : Grading Program
Note: Use the concepts learnt so far. Description: Write a program to take the marks of 5 subjects as input from the user, and print a result card. The result card should contain the following:

1. The date and name, fathername, roll number of the student
2. Total Marks, obtained marks and grades secured in each subject.
3. The total marks (all subjects), Marks Obtained and Grades secured overall (all subjects).
4. Make sure the handle edge cases. (e.g. if the user enters a string instead of a number, or a number greater than 100, etc.)
5. The program should be dynamic i.e. it should ask for the student's name, father's name, roll number, and marks of each subject and perform the desired operations based on the provided input.

hint:

1. use the input() function to take input from the user.
2. use if-else statements to handle edge cases.
3. use f-string to format the output of the program i.e. result card.
4. Subjects: Math, Physics, Chemistry, Computer Science, English. Each subjects would have total marks of 100.
```

```python
from datetime import datetime

def get_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'

def get_marks(subject):
    while True:
        try:
            marks = float(input(f"Enter marks for {subject}: "))
            if 0 <= marks <= 100:
                return marks
            else:
                print("Marks should be between 0 and 100. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def main():
    # Getting student details
    name = input("Enter the student's name: ")
    father_name = input("Enter the father's name: ")
    roll_number = input("Enter the roll number: ")
    
    # Initializing subject details
    subjects = ["Math", "Physics", "Chemistry", "Computer Science", "English"]
    total_marks = 500
    obtained_marks = 0
    subject_details = []
    
    # Getting marks for each subject
    for subject in subjects:
        marks = get_marks(subject)
        obtained_marks += marks
        grade = get_grade(marks)
        subject_details.append((subject, marks, grade))
    
    # Calculating overall grade
    overall_grade = get_grade(obtained_marks / len(subjects))
    
    # Printing the result card
    print("\n--- Result Card ---")
    print(f"Date: {datetime.now().strftime('%Y-%m-%d')}")
    print(f"Name: {name}")
    print(f"Father's Name: {father_name}")
    print(f"Roll Number: {roll_number}")
    print("\nSubject-wise Details:")
    for subject, marks, grade in subject_details:
        print(f"{subject}: Marks Obtained = {marks}, Grade = {grade}")
    
    print(f"\nTotal Marks: {total_marks}")
    print(f"Marks Obtained: {obtained_marks}")
    print(f"Overall Grade: {overall_grade}")

if __name__ == "__main__":
    main()
```
